## Preparación y Carga de datos

### Librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_iris
```

### Carga de datos

```python
data=load_iris()
data.keys()
```

```python
data['target_names']
```

```python
data['data']
```

```python
variables=pd.DataFrame(data['data'], columns=data['feature_names'])
variables.head()
```

```python
variables['clase']=data['target']
variables
```

```python
data['target_names']
```

```python
# otra forma de hacer un diccionario dodne se le da los valores 0 al primer elemento , 1 al segundo ....
#dict(zip(range(len(data['target_names'])),data['target_names']))
# pero como el dic es pequeño usremos la vorma de abajo 
```

```python
variables['clase'].replace({0:'setosa',1:'versicolor',2:'virginica'},inplace=True)
variables.to_parquet('iris.paquet',index=False)
```

```python
df=pd.read_parquet('iris.paquet')
df
```

## Análisis Diagnósitco

### Análisis de la separabilidad de flores

```python
plt.figure(figsize=(12,5))

# relacion de petalo

plt.subplot(1,2,1)
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='clase')
plt.title('Relacion entre longitud y ancho de pétalo')

# relacion de sepal

plt.subplot(1,2,2)
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='clase')
plt.title('Relacion entre longitud y ancho de sépalo')

plt.tight_layout()
plt.show();

```

```python
# Análisis de ratios

df['ratio_petal_sepal_length']=df['petal length (cm)']/df['petal width (cm)']
df['ratio_petal_sepal_width']=df['sepal length (cm)']/df['sepal width (cm)']

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxenplot(data=df,x'clase',y='ratio_petal_sepal_length')
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxenplot(data=df,x'clase',y='ratio_petal_sepal_width')
plt.tight_layout()
plt.show();
```
