## Ejercicio 1: Ortogonalidad

Definir una funci칩n en python que detecte si dos vectores son ortogonales

```python
import numpy as np

#Para verificar si dos vectores son ortogonales, debes calcular su producto escalar. 
# Dos vectores son ortogonales si y solo si su producto escalar es igual a cero.

def son_ortogonales(u,v):
    producto_escalar= np.dot(u,v)
    if producto_escalar==0:
        return True
    else:
        return False
    
```

```python
son_ortogonales([1,2,3],[3,-6,1])
```

```python
son_ortogonales([-2,-2],[2,-2])
```

## Ejercicio 2: Desplazamiento

Definir una funci칩n en python que reciba como par치metro las cuatro esquinas de un cuadril치tero y las desplace 5 unidades hacia arriba y 7 unidades a la izquierda

---

```python
# Este es el modo en el que yo pense principalmente , la forma que se me pedia esta mas abajo

def desplazar_puntos_1(array_1):
  r=np.array([(-7,5)])
  return array_1[0]+r[0],array_1[1]+r[0],array_1[2]+r[0],array_1[3]+r[0]
  
```

```python
import matplotlib.pyplot as plt

u=np.array([(1,1),(3,1),(2,3),(4,3)])

x_coords, y_coords = zip(u[0],u[1],u[2],u[3])
plt.scatter(x_coords, y_coords, color=["r","b","y","c"])
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
```

```python
g=desplazar_puntos_1(u)
x_coords, y_coords = zip(g[0],g[1],g[2],g[3])
plt.scatter(x_coords, y_coords, color=["r","b","y","c"])
plt.axis([0, -9, 4, 9])
plt.grid()
plt.show()
```

```python
def desplazar_puntos(a,b,c,d):
  r=np.array([-7,5])
  return r+a,r+b,r+c,r+d
  
```

Ejecutar la siguiente celda descomentando las lineas en verde para verificar su funcionamiento.

```python
import matplotlib.pyplot as plt
a = np.array([1,1])
b = np.array([1,2])
c = np.array([2,3])
d = np.array([8,4])

x_coords, y_coords = zip(a,b,c,d,a)
plt.plot(x_coords, y_coords, "c--", x_coords, y_coords, "co")

a_,b_,c_,d_ = desplazar_puntos(a,b,c,d)
x_coords_b, y_coords_b = zip(a_,b_,c_,d_,a_)
plt.plot(x_coords_b, y_coords_b, "b-", x_coords_b, y_coords_b, "bo")

plt.show()
```
