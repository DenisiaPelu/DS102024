### Paso 2. Importar el conjunto de datos de esta [direcci√≥n](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

```python
df=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv")
df.head()
```

### Paso 3. Asignarlo a una variable llamada tips

```python
tips=df.copy()
```

### Paso 4. Trazar el histograma de la columna ``total_bill``

```python
tips.columns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], bins=20, kde=True, color='skyblue')
            
    # Agregar detalles al gr√°fico
plt.title(f'Histograma de {"total_bill"}', fontsize=14)
plt.xlabel('total_bill', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Paso 5. Cree un gr√°fico de dispersi√≥n que presente la relaci√≥n entre ``total_bill`` y ``tip``

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', alpha=0.6)
plt.title('grafico de dispersion entre  Score vs. members')
plt.xlabel('rating')
plt.ylabel('members')
plt.show()
```

### Paso 6.  Crear una imagen con la relaci√≥n de `total_bill`, `tip` y `size`.
#### Sugerencia: es s√≥lo una funci√≥n.

```python
df = sns.load_dataset('tips')

sns.scatterplot(data=df, x='total_bill', y='tip', hue='size', palette='viridis', size='size', sizes=(20, 200))

plt.title('Relaci√≥n entre Total Bill, Tip y Size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```

```python
df = sns.load_dataset('tips')
cols = ['total_bill', 'tip', 'size']
sns.pairplot(df[cols], kind='scatter')

plt.show()
```

### Paso 7. Presente la relaci√≥n entre ``days`` y ``total_bill``

```python
tips.columns
```

```python
sns.scatterplot(data=tips, x='day', y='total_bill')

plt.title('Relaci√≥n entre Day y Total Bill')
plt.xlabel('D√≠a de la semana')
plt.ylabel('Total Bill')

plt.show()
```

### Paso 8. Crear un gr√°fico de dispersi√≥n con el ``day`` como eje y y ``tips`` como eje x, diferenciar los puntos por ``sex``.

```python
sns.scatterplot(data=df, x='tip', y='day', hue='sex', palette='coolwarm')

plt.title('Relaci√≥n entre Tip y Day, diferenciando por Sexo')
plt.xlabel('Tip')
plt.ylabel('D√≠a de la Semana')

plt.legend(title='Sexo')

# Mostrar el gr√°fico
plt.show()
```

### Paso 9.  Cree una box plot presentaci√≥n del ``total_bill`` por ``day`` diferetiation el tiempo ``(Dinner or Lunch)``

```python
sns.boxplot(data=df, x='day', y='total_bill', hue='time', palette='Set2')

plt.title('Distribuci√≥n de Total Bill por D√≠a de la Semana, Diferenciado por Tipo de Comida')
plt.xlabel('D√≠a de la Semana')
plt.ylabel('Total Bill')

# Mostrar el gr√°fico
plt.show()
```

### Paso 10. Cree dos histogramas del valor del tip basados en Dinner y Lunch. Deben estar uno al lado del otro.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.histplot(data=tips[tips['time'] == 'Dinner'], x='tip', kde=True, color='blue', ax=axes[0])
axes[0].set_title('Distribuci√≥n de Propina para Dinner')
axes[0].set_xlabel('Tip')
axes[0].set_ylabel('Frecuencia')

sns.histplot(data=tips[tips['time'] == 'Lunch'], x='tip', kde=True, color='green', ax=axes[1])
axes[1].set_title('Distribuci√≥n de Propina para Lunch')
axes[1].set_xlabel('Tip')
axes[1].set_ylabel('Frecuencia')

# Ajustar espacio entre los subgr√°ficos
plt.tight_layout()

plt.show()
```

### Paso 11. Cree dos gr√°ficos de dispersi√≥n, uno para los hombres y otro para las mujeres, que presenten la relaci√≥n entre el valor ``total_bill`` y la ``tip``, diferenciando entre ``smoker`` y ``no smoker``.
### Deben estar uno al lado del otro.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.scatterplot(data=tips[tips['sex'] == 'Male'], x='total_bill', y='tip', hue='smoker', palette='coolwarm', ax=axes[0])
axes[0].set_title('Relaci√≥n entre Total Bill y Tip para Hombres')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')

sns.scatterplot(data=tips[tips['sex'] == 'Female'], x='total_bill', y='tip', hue='smoker', palette='coolwarm', ax=axes[1])
axes[1].set_title('Relaci√≥n entre Total Bill y Tip para Mujeres')
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Tip')

plt.tight_layout()

plt.show()
```
