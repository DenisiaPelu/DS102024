### Step 19. Veamos la función melt() de pandas. Melt toma los valores de varias columnas y los junta todos en una misma columna nueva.
 Dado un DataFrame, [melt](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.melt.html) recibe varios parámetros:
 * qué columnas del DataFrame se quedan igual
 * qué nombres de columnas del dataframe pasan a ser una columna nueva
 * cómo queremos llamar esas nuevas columnas

Primero recuerda el Dataframe `df_grass_fire_water` con las primeras 5 filas

```python
df_copia['Type 1'].unique()
```

```python
df_copia_grass_fire_water=df_copia
```

```python
df_copia_grass_fire_water.columns
```

- Las columnas ['Name', 'Type 1'] se quedan igual.
- las columnas ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'] se fusionan.
- La nueva columna para los valores fusionados es 'Stat' 
- Lanueva coluna para los valores es  'Value'

```python
df_melte = df_copia_grass_fire_water.melt(id_vars=['Name', 'Type 1'],value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'],var_name='Stat',value_name='Value')

df_melte.head()
```

### Step 20. Queremos dejar igual las variables 'Name' y 'Type 1'
Queremos añadir una columa nueva cuyos valores sean las columnas que queremos apilar y que se llame 'New Column'
    
Escribimos `df_grass_fire_water.melt(id_vars=['Name', 'Type 1'], value_vars=[...], var_name='New Column')`

```python
df_melte_2 = df_copia_grass_fire_water.melt(id_vars=['Name', 'Type 1'],value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'],var_name='Melted',value_name='Value')
df_melte_2.head()
```

### Step 21. Pintamos el swarmplot con el melted y entendemos la información

```python
sns.swarmplot(x='Melted',y='Value',data=df_melte_2,hue='Type 1',palette=color,dodge=True)
plt.show();
```

### Step 24. Pinta el histograma y la densidad de probabilidad del nivel de defensa usando el DataFrame original

```python
def generar_histogramas( df, bins=20):
        """
        Genera histogramas para todas las columnas numéricas de un DataFrame.
        
        :param df: DataFrame con los datos.
        :param bins: Número de bins para los histogramas.
        """
        columnas_numericas = df.select_dtypes(include=['number']).columns
        
        # Crear histogramas para cada columna
        for column in columnas_numericas:
            plt.figure(figsize=(10, 6))
            sns.histplot(df[column], bins=bins, kde=True, color='skyblue')
            
            # Agregar detalles al gráfico
            plt.title(f'Histograma de {column}', fontsize=14)
            plt.xlabel(column, fontsize=12)
            plt.ylabel('Frecuencia', fontsize=12)
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.show()
```

```python
generar_histogramas(df)
```

### Step 26. Usa catplot con el DataFrame original. 
Muestra el valor de Defense según el 'Type 1', muestra un subplot para legendarios y otro para no legendariosy divide los colores según el Type 2

```python
sns.catplot(
    data=df,    
    x='Type 1', 
    y='Defense', 
    hue='Type 2',        
    col='Legendary',     
    kind='strip',          # Tipo de gráfico (puede ser 'strip', 'swarm', 'box', etc.)
    palette='tab20',  
    height=6,            # Altura de cada gráfico
    aspect=1.5           # Relación de aspecto (ancho)
)

# Ajustar etiquetas y rotación para mejorar la legibilidad
plt.subplots_adjust(top=0.9)  # Ajustar el espacio superior
plt.suptitle("Defensa según Tipo 1 y Legendary", fontsize=16)  # Título general
plt.xticks(rotation=45, fontsize=10)

# Mostrar el gráfico
plt.show();
```

### Step 27. Muestra una función distribución de probabilidad con kdeplot (curvas de nivel) de la defensa respecto al ataque.


```python
plt.figure(figsize=(10, 8))

sns.kdeplot(
    data=df,      
    x='Attack', 
    y='Defense',
    cmap='Blues',
    fill=True, 
    levels=20, 
    thresh=0 
)

plt.title('Distribución de Defensa respecto al Ataque', fontsize=16)
plt.xlabel('Ataque', fontsize=14)
plt.ylabel('Defensa', fontsize=14)

plt.show();
```

### Step 29. Pinta una distribución de probabilidad conjunta, pintando la defensa respecto al ataque del DataFrame original

```python
sns.jointplot(
    data=df, 
    x='Attack', 
    y='Defense',
    kind='kde', 
    fill=True,
    cmap='coolwarm', 
    space=0                
)

plt.suptitle('Distribución Conjunta: Defensa vs. Ataque', fontsize=16, y=1.02)

plt.show();
```
