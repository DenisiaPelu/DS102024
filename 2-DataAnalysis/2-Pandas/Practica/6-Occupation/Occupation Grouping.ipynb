### Paso 6. Para cada ocupación, calcule las edades mínima y máxima

### Paso 7. Para cada combinación de ocupación y sexo, calcule la edad media

```python
usuario.columns
```

```python
usuario.groupby(['occupation','gender'])['age'].mean()
```

- otra manera de hacer

```python
usuario.pivot_table(
    values='age',
    index='occupation',
    columns='gender',
    aggfunc='mean'
)
```

### Paso 8.  Para cada ocupación, presente el porcentaje de mujeres y hombres

```python
df_1=usuario.groupby(['occupation','gender']).size()
df_2=df_1.groupby(level=0).apply(lambda x: 100*x/x.sum())
df_2=df_2.rename('porcentaje')
```
