```python
import hashlib
import requests
import datetime
import pandas as pd
import json
from IPython.display import Image, HTML
```

```python
def hash_params(timestamp,priv_key,pub_key):
    """ Marvel API requires server side API calls to include
    md5 hash of timestamp + public key + private key """

    hash_md5 = hashlib.md5()
    hash_md5.update(f'{timestamp}{priv_key}{pub_key}'.encode('utf-8'))
    hashed_params = hash_md5.hexdigest()

    return hashed_params
```

```python
timestamp = datetime.datetime.now().strftime('%Y-%m-%d%H:%M:%S')

pub_key = '13c93276c28278ebef4f61fd9a3c7bfa'
priv_key = '5fdbef310d1949f0e24badaa05806064e8bddcc7'

params = {'ts': timestamp, 
        'apikey': pub_key, 
        'hash': hash_params(timestamp,priv_key,pub_key),
        'nameStartsWith':"D",
        'offset':0,
        'limit': 100
        
        };

url = 'http://gateway.marvel.com/v1/public/characters'

res = requests.get(url,params=params)

```

```python
data=res.json()
```

```python
# Crear una lista de diccionarios con el nombre y la URL 
marvel_lis = [{"id": D["id"], "name": D["name"],"picture_urls": f"{D['thumbnail']['path']}.{D['thumbnail']['extension']}" } for D in data["data"]["results"]]

df_marvel = pd.DataFrame(marvel_lis)
df_marvel
```

```python
def display_image(url):
    return f'<img src="{url}" width="100" />'
df_marvel['Image'] = df_marvel['picture_urls'].apply(display_image)
# Mostrar como HTML en Jupyter Notebook
HTML(df_marvel.to_html(escape=False))
```
