## Creamos el modelo

- a partir de ahora hacemos experimentos de ensayo y error.
- si supieramos estadistica sabriamos que tipo deberiamos usar

```python
X = np.array(dataframe.drop(columns=['clase']))
y = np.array(dataframe['clase'])
```

que tipo de dato es nuestra targer Clase ( la trataremos como categorica ) es modelo supervisado por tener 3 categorias (0,1,2) vamos a clasificar ( Casificacion)

```python
# tipo de algoritmo
from sklearn.linear_model import LogisticRegression
```

- faltaria separar en try y text ... pero hay muy pocos datos.

```python
# 1000 interaciones tb aumente el tiempo de comput, pero tb es mas precisoi en el entrenamiento , tb sobre entrenar es mala (over fili).
model = LogisticRegression(max_iter=1000) 
model.fit(X, y)
```

```python
predictions = model.predict(X)
print(predictions)
```

```python
# la  probabilidad de que tipo es 
predicions_proba = model.predict_proba(X)
print(np.round(np.array(predicions_proba), 2))
```

```python
model.score(X, y)
```

```python
model.classes_
```

# Adicional: Validación del Modelo

- vamos a dividir en Train y en Test

```python
validation_size = 0.20
seed = 7
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,
                                                                    y,
                                                                    test_size=validation_size,
                                                                    random_state=seed)
```

```python
# name='Logistic Regression'
# kfold = model_selection.KFold(n_splits=10) #Parte los datos en 10 trozos para usar validación cruzada / cross validation
# cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')

# msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
# print(cv_results)
# print(msg)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
print(accuracy_score(Y_test, predictions))
```

```python
acierto = accuracy_score(Y_test, predictions)

error = 1 - acierto
print("Acierto:", round(acierto*100, 2), "%")
print("Error:", round(error*100, 2), "%")
```
